### 4.1 The Bug Collector Problem
A bug collector collects bugs every day for five days. Write a program that keeps a running total
of the number of bugs collected during the five days. The loop should ask for the number of
bugs collected for each day, and when the loop is finished, the program should display the total
number of bugs collected.

```python
day_count = 1
total_bugs = 0

# while (day_count < 6):  # ok
# while (day_count <= 5): # ok
# while (day_count < 5):  # not ok; One-off defect
    
while (day_count <= 5):    
  print("Day Number: ", day_count)
  bug_count = input("Enter the bugs you collected: ")
  bug_count = int(bug_count)
  total_bugs = total_bugs + bug_count
  day_count = day_count + 1
  
# output
print("Total number of bugs collected: ", total_bugs)
```

### 4.2. Calories Burned
Running on a particular treadmill you burn 4.2 calories per minute. Write a program that uses a
loop to display the number of calories burned after 10, 15, 20, 25, and 30 minutes.

```python
# declare and initialize variables
CAL_PER_MINUTE = 4.2
starting_minute = 10

# setup for loop
while (starting_minute < 31):
  calories_burned = starting_minute * CAL_PER_MINUTE
  print("Number of calories burned after", starting_minute, "minutes: ", calories_burned)
  starting_minute = starting_minute + 5

# output
print("Done!")

```

### 4.7. Pennies for Pay
Write a program that calculates the amount of money a person would earn over a period of time
if his or her salary is one penny the first day, two pennies the second day, and continues to
double each day. The program should ask the user for the number of days. Display a table
showing what the salary was for each day, then show the total pay at the end of the period. The
output should be displayed in a dollar amount, not the number of pennies.


```python
# get the input from the user
max_day = input("Enter the number of days: ")
max_day = int(max_day)


# initialize the variables
day_num = 1   # we need this to check the loop condition
pay_for_the_day = 1  # for keeping track of the pay for a given day
total_pay = 0  # need this to print the final output

# setup the while loop
while (day_num <= max_day):
    if day_num == 1 :
       pay_for_the_day = 1 
    else :
        pay_for_the_day = pay_for_the_day * 2
    print('day', day_num, ' = ' , pay_for_the_day, ' pennies')
    
    day_num = day_num + 1
    total_pay = total_pay + pay_for_the_day
  
# print the final output
print('The total pay in pennies' , total_pay)
print('The total pay in dollars' , total_pay/100)
```

## WHILE loop - sentinel loop

A sentinel loop continues to process data until reaching a special value that signals the end. 

The special value is called the sentinel.

Any value may be chosen for the sentinel. The only restriction is that it be distinguishable from actual data values. 

The sentinel is not processed as part of the data.


```python
#@title WHILE (sentinel example 1)

# Initialize the variables
number_of_students = 0
age = 1  # to get the loop going
total_age = 0

# setup the while loop (sentinel is age = 0)
while (age > 0):
  # sentinel value is - any negative number
  age = input("Enter the age? (Must be a +ve number; OR enter 0 to signal the end) ")
  age = int(age)
  if (age != 0):
    total_age = total_age + age
    number_of_students = number_of_students + 1

# find the average and display the output
average = total_age / number_of_students
print("Number of Students: ", number_of_students)
print("Total Age of Students: ", total_age)
print("Average age: ", average)
```

## break statement

```python
#@title I don't like 13! Stop the train!
# Note:  Execute this program multiple times
# Print 10 random numbers between 1 and 25. 
# But if the random number generated is 13, just stop the loop

import random

loop_counter = 1
while (loop_counter <= 10):
  random_x = random.randint(1,25)
  if random_x == 13:
    print("13 is the random number! I am stopping!")
    break

  print(loop_counter, "Random number: ", random_x)
  loop_counter = loop_counter + 1
```

```python
#@title while True and break
# break, sentinel value and True are typically used together.
# Keep on running the loop (while True)
# Break out of the loop when some 'sentinel' value is reached


# Initialize the variables
number_of_students = 0
total_age = 0

# setup the while loop (sentinel is age = 0)
while True:
  # sentinel value is - any negative number
  age = input("Enter the age? (Must be a +ve number; OR enter 0 to signal the end) ")
  age = int(age)

  # if age is 0, break out of the loop
  if (age == 0):
    break

  # if age is not 0, proceed in the normal path
  total_age = total_age + age
  number_of_students = number_of_students + 1

# find the average and display the output
average = total_age / number_of_students
print("Number of Students: ", number_of_students)
print("Total Age of Students: ", total_age)
print("Average age: ", average)
```

## continue statement

```python
#@title I don't like 13! Skip it and keep moving!
# Note:  Execute this program multiple times
# Print 10 random numbers between 1 and 25. 
# But if the random number generated is 13, skip that number

import random

loop_counter = 1
while (loop_counter <= 10):
  random_x = random.randint(1,25)
  
  if random_x == 13:
    print("13 is skipped! But I am moving ahead with the rest!")
    continue

  print(loop_counter, "Random number: ", random_x)
  loop_counter = loop_counter + 1
```

```python
#@title 5.3. Assignment continue and break (an exercise)

'''
Run a While loop infinitely (using while True).

Inside the loop, generate two random numbers between (1 and 10)

If both the random numbers are equal, then the loop should skip those and continue with the next iteration.

If the sum of the random numbers is 10, then the loop should exit.

Otherwise, the loop will print the two random numbers and their sum.


For example, the typicaly output of the program can be as follows. (your output may be different)

Number 1: 8       Number 2: 3         Sum: 11
Number 1: 5       Number 2: 7         Sum: 12
Number 1: 4       Number 2: 6         Sum: 10
Exiting the program as the sum is 10

'''
```

```python
#@title Solution to Assignment 5.3
import random

while True:
  random_1 = random.randint(1,10)
  random_2 = random.randint(1,10)

  if (random_1 == random_2):
    continue
  
  sum = random_1 + random_2

  if (sum == 10):
    print("Exiting the program as the sum is 10")
    break

  print("Number 1: ", random_1, "      Number 2: ", random_2, "      Sum: ", sum)
```

## Case Study: While, True, Break, Continue
See google class for the case study

```python
# prompt the user
print('Enter a series of +ve numbers (or “quit” to stop):')
print('Note that –ve and 0 will be ignored.')


# initialize the output variables
total = 0
min = -1
max = -1 

# setup the while loop
while True:
    number = input()
    
    # break on the sentinel
    if (number == 'q'):
        break
    
    # convert str to int
    number = int(number)
    
    # ignore -ve and 0
    if (number <= 0):
        continue
    
    # update the total
    total = total + number
    
    # update max
    if (max == -1):
        max = number
    elif (number > max):
        max = number
    
    # update min  
    if (min == -1):
        min = number
    elif (number < min):
        min = number
        
    
# print the outputs

print('Minimum = ', min)
print('Maximum = ', max)
print('Sum = ', total)
```
