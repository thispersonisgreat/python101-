```python
#@title read_gas_prices()
'''
Method Name: read_gas_prices
Input: file name
Returns: None
What global variables are updated? prices_master_list


This function takes in a gas prices file name and creates a prices_master_list
'''
def read_gas_prices(file_name):
  DataFile = open("gasprices.txt", "r")
  data = DataFile.read()
  prices_master_list = data.split("\n")
  return prices_master_list
```

```python
from typing_extensions import dataclass_transform
#@title  get_average_price_per_year()

'''
Method Name: get_average_price_per_year
Input: None (it relies on the global variable prices_master_list)
Returns: None
What global variables are updated? None
Displays:

Year      Average Price
1993     1.07
1994     1.09


This function takes in a gas prices file name and creates a prices_master_list
'''

# DataFile = open("gasprices.txt", "r")
# data = DataFile.readlines()
# for line in data:
#   print(line.split(":")[1])



def get_average_price_per_year():
    # set up a for loop for years

    DataFile = open("gasprices.txt", "r")
    data = DataFile.readlines()

    for year in range(1993, 2013):
      count = 0
      add = 0
      for line in data:
        dataYear = line.split(":")[0].split("-")[2]
        dataPrice = float(line.split(":")[1])
        if str(dataYear) == str(year):
          count = count + 1
          add = add + int(dataPrice)
      print("year:",year," average:",round(add/count, 2))
      
    # you need to read the prices_master_list from beginning to end
    # if the year matches the current year, add the current price to the total
    # and print the year and average


get_average_price_per_year()
```
